```python
# Load packages
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date, time
import os 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Get current directory
path = os.getcwd()
data_path = path + '/Data/'
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'yellow', 'cyan', 'magenta', 'olive', 'maroon', 'navy', 'aqua', 'teal']
# Find datafile
file = 'Rphenograph_output_275clusters_k250_16ct_fractions.csv'
pi = pd.read_excel('Data/raw/LUAD Clinical Data.xlsx')

df = pd.read_csv(f"{data_path}/phenograph_output/{file}", index_col=0)
```

```python
pi
```

```python
def above_5(x):
    if x > 5:
        return 1
    else:
        return 0

pi['surv_5y'] = pi['Survival or loss to follow-up (years)'].apply(lambda x: above_5(x))
s5 = pi[['Key','surv_5y']].rename(columns={"Key":"Patient_ID"})
s5
df = pd.merge(df, s5, on='Patient_ID', how='inner')
df
```

```python
high_surv = df[df['surv_5y'] == 1]
low_surv = df[df['surv_5y'] == 0]
low_surv
```

```python
low_surv['Patient_ID'].nunique()
```

```python
# Group by cluster and count unique patients
cluster_counts = low_surv.groupby('cluster')['Patient_ID'].nunique().reset_index()

# Sort clusters based on the count of unique patients
cluster_counts = cluster_counts.sort_values(by='Patient_ID', ascending=False)
# Optional: Create a new DataFrame to store the cluster rankings
cluster_rankings = pd.DataFrame({
    'cluster': cluster_counts['cluster'],
    'unique_patients_count': cluster_counts['Patient_ID'],
    'rank': range(1, len(cluster_counts) + 1)
})
cluster_rankings
```

```python
plt.scatter(cluster_rankings['rank'], cluster_rankings['unique_patients_count'], s=1)
plt.title('Low survival cohort the number of unique patients clusters occur in against their rank')
plt.xlabel('Rank')
plt.ylabel('Number of unique patient occurences')
```

```python
# Count the number of cells in each group
group_counts = df['cluster'].value_counts()

# Get the groups with less than 100 cells
small_groups = group_counts[group_counts < 10000].index

# Merge the small groups
df.loc[df['cluster'].isin(small_groups), 'cluster'] = 'Other'
```

```python
# Group by 'group' and 'cell_type', and count the number of occurrences
grouped = df.groupby(['cluster', 'source_cluster']).size().unstack(fill_value=0)

# Plot the makeup of groups colored by cell type
grouped.plot(kind='bar', stacked=True, colormap='tab20')

# Add labels and title
plt.xlabel('Group')
plt.ylabel('Count')
plt.title('Makeup of clusters colored by Cell Type')
plt.legend(title='Cell Type', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
# Code to merge low count clusters
# Count the number of cells in each group
group_counts = high_surv['cluster'].value_counts()

# Get the groups with less than 100 cells
small_groups = group_counts[group_counts < 5000].index

# Merge the small groups
high_surv.loc[high_surv['cluster'].isin(small_groups), 'cluster'] = 'Other'
```

```python
high_surv = df[df['surv_5y'] == 1]
low_surv = df[df['surv_5y'] == 0]
low_surv
```

```python
# Grouping by 'cluster' and 'celltype' and getting counts
result = high_surv.groupby(['cluster', 'source_cluster']).size().reset_index(name='Count')

# Pivot the table to get totals per cluster and cell type
result = result.pivot_table(index='source_cluster', columns='cluster', values='Count', fill_value=0)

# Add totals for each cell type and each cluster
result['Total_Cell_Type'] = result.sum(axis=1)
result.loc['Total_Cells'] = result.sum()
```

```python
result_percentage = result.div(result.loc['Total_Cells'], axis=1) * 100

# Plotting
ax = result_percentage.iloc[:-1,:-1].T.plot(kind='bar', stacked=True, figsize=(10, 6),width=0.5, color=colors)
ax.set_ylabel('Percentage')
ax.set_xlabel('Cluster number')
ax.set_title('High survival cohort stacked Barplot of Cell Types in Clusters')

# Annotate total amount of cells above each bar
for i, total in enumerate(result.loc['Total_Cells'][:-1]):  # Exclude the 'Total_Cells' row
    ax.annotate(f'{total}', (i, 100), xytext=(0, 5), textcoords='offset points', ha='center', rotation=15, fontsize=5)

# Show plot
plt.legend(title='Cell Type', bbox_to_anchor=(1, 1))
plt.savefig(f'{data_path}plots/highsurv_stacked_barplot.png', bbox_inches='tight')

plt.show()
```
