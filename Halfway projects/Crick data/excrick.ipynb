```python
relevant_columns = ['Location_Center_X', 'Location_Center_Y', 'ExpGroup',  'MouseID','ROI_ID', 'cluster_num', 'Unique_ROI_ID']
```

```python
groups = ['MOC1_MOCAF', 'MOC1_WT', 'MOC2_CCR2KO', 'MOC2_WT']
group_dict = {}
for group in groups:
    temp = df2[df2['ExpGroup'] == group][relevant_columns]
    # Give cellnumber to cells
    temp['CellID'] = df.groupby('Unique_ROI_ID').cumcount() + 1
    group_dict[group] = temp.drop(columns=['ExpGroup'], inplace=False)
    group_dict[group].to_csv(f'Data/celldata/{group}_celldata_{date}.csv', index=False)

group_dict['MOC1_MOCAF']
```
